```python
import matplotlib.pyplot as plt
from modem_plotter import ModemPlotter

plotter = ModemPlotter('output/NUAE_NLCG_071.csv', {
    'x_range_km': (-35, 35), 'y_range_km': (-30, 30), 'z_range_km': (0, 45), 
    'cbar_y_position': 0.2, 'wspace': 0.02, 'hspace': 0.08,
    'dpi': 600, 'save_dpi': 600,
    'cmap': 'turbo_r', 'vmin': 0.0, 'vmax': 5.0, 'verbose': True, 'output_dir': 'output', 
})

# East-West Vertical Slice
x_positions = [-20.0, -10.0, -6.0, 0.0, 6.0, 10.0, 20.0]  # in km
plots_x = [{'type': 'vertical', 
            'orientation': 'x', 
            'position': x*1000,
            'title': f'E = {x} km'} for x in x_positions]

fig1 = plotter.plot_multiple(plots_x, ncols=3, figsize=(10, 11.5))
if fig1:
    plotter.save_figure(fig1, 'vertical_slices_x_uae.png')
    plt.show()

# North-South Vertical Slices
y_positions = [-20.0, -10.0, -6.0, 0.0, 6.0, 10.0, 20.0]  
plots_y = [{'type': 'vertical', 
            'orientation': 'y', 
            'position': y*1000, 
            'title': f'N = {y} km'} for y in y_positions]

fig2 = plotter.plot_multiple(plots_y, ncols=3, figsize=(10, 10))
if fig2:
    plotter.save_figure(fig2, 'vertical_slices_y_uae.png')
    plt.show()
```
